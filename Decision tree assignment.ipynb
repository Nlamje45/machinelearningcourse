```python
import numpy as np
import pandas as pd
data=pd.read_csv(r"C:\Users\ADMIN\Downloads\titanic.csv")
print(data)
```

```python
data.columns
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.cross_validation import train_test_split
```

```python
data1=data[['pclass','sex','age','sibsp','parch','fare','survived']]
data2=data1.dropna(axis=0)
print(data2)
```

```python
x=data2[['pclass','sex','age','sibsp','parch','fare']]
y=data2['survived']
print(x)
```

```python
x=pd.get_dummies(x,drop_first=True)
x
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=10)
```

```python
m1=DecisionTreeClassifier()
m1.fit(xtrain,ytrain)
p=m1.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
```

```python
c=accuracy_score(p,ytest)
c
```
